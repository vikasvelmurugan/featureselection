<b> METHOD 2: SELECTKBEST FEATURES </b>

```python
X=data.iloc[:,:-1]
y=data['price_range']
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

bestFeatures= SelectKBest(score_func=chi2, k=10)
model=bestFeatures.fit(X,y)
```

```python
model.scores_
```

```python
dataScore=pd.DataFrame(model.scores_)
dataCols=pd.DataFrame(X.columns)
featureScores= pd.concat([dataCols,dataScore], axis=1)
featureScores.columns=["Features", "Scores"]
featureScores
```

<b> Printing the top 10 highly correlated features <b>

```python
featureScores.nlargest(10,"Scores")
```

```python
featureScores.nlargest(10, "Scores").plot(kind="bar")
featureScores.
plt.show()
```
